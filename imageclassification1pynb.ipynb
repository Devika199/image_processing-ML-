```python
catpath=os.path.join("/content/drive/MyDrive/Data","Cat")# join-it is used to join 2 paths like cat and data
for img in os.listdir(catpath):# to display all images in cat
  print(img)
```

```python
# to convert this into numerical.we save it as in array
flat_data_arr=[] #used to store features.to convert it into nu
target_arr=[]# this is to select if it is dog or cat ie to label if it is cat or dog
categories=["Cat","Dog"]# cat -0, dog-1
```

```python
data_dir=("/content/drive/MyDrive/Data")
for i in categories:
  print("loading ...category",i)
  path=os.path.join(data_dir,i)# to join 2 paths 1st part is datapath+catpath .here we get path of cat
  for img in os.listdir(path):
    img_array=imread(os.path.join(path,img))# to read an an image
    img_resized=resize(img_array,(150,150,3))# to resize image,here we get 3d
    flat_data_arr.append(img_resized.flatten())#flatten is used to convert 3d,2d to 1d
    target_arr.append(categories.index(i))# cat-0,dog-1
  print('loaded',i)   

```

```python
#jst creating dataframe
#intially we had 126 values but using flatten we get 6750
df=pd.DataFrame(flat_data_arr)
df
```

```python
df["target"]=target_arr
df
```

```python
x=df.iloc[:,:-2]
x
```

```python
y=df.iloc[:,-2]
y
```

```python
df.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=6)

```

```python
from sklearn.svm import SVC
sv=SVC()#(kernel="linear")
sv.fit(x_train,y_train)
ypred=sv.predict(x_test)
ypred
```

```python
from sklearn.metrics import classification_report,accuracy_score,ConfusionMatrixDisplay
print(accuracy_score(y_test,ypred))
```

```python
print(ConfusionMatrixDisplay.from_predictions(y_test,ypred))
```
